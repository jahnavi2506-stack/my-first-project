```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

root = Node(int(input("Root: ")))
root.left = Node(int(input("Left: ")))
root.right = Node(int(input("Right: ")))

print("\n")
print("      ", root.data)
print("     /   \\")
print("   ", root.left.data, "   ", root.right.data)
```

```python
#decimal to binary
n = int(input("Enter a decimal number: "))
binary = ""
while n > 0:
    binary = str(n % 2) + binary
    n = n // 2
print("Binary:", binary)    
```

```python
#decimal to binary using append,slicing & print in list
n = 13
binary = []
while n > 0:
    remainder = n % 2
    binary.append(remainder)
    n = n // 2
binary = binary[::-1]
print(binary)
```

```python
#binary to decimal
binary = input("Enter a binary number:")
binary = binary[::-1]
decimal = 0
for i in range(len(binary)):
    decimal += int(binary[i]) * (2**i)
print(decimal)    
```

```python
#decimal to octal
n = int(input('Enter a decimal number:'))
octal = []
while n > 0:
    octal.append(str(n % 8))
    n = n // 8
octal = octal[::-1]
print("".join(octal))
    
```

```python
#octal to decimal
octal = input("Enter an octal number: ")
octal = octal[::-1]
decimal = 0
for i in range(len(octal)):
    decimal += int(octal[i]) * (8 ** i)
print(decimal)
```

```python
#binary to hexa decimal
binary = input("Enter a binary number: ")
# Binary → Decimal
binary_rev = binary[::-1]
decimal = 0
for i in range(len(binary_rev)):
    decimal += int(binary_rev[i]) * (2 ** i)
# Decimal → Hexadecimal
hex_digits = "0123456789ABCDEF"
hexa = []
while decimal > 0:
    hexa.append(hex_digits[decimal % 16])
    decimal = decimal // 16
hexa = hexa[::-1]
print("".join(hexa))
```

```python
#decimal to hexa decimal 
n = int(input("Enter a decimal number: "))
hex_digits = "0123456789ABCDEF"
hexa = []
while n > 0:
    remainder = n % 16
    hexa.append(hex_digits[remainder])
    n = n // 16
hexa = hexa[::-1]
print("".join(hexa))
```

```python
# hexadecimal to decimal
hexa = input("Enter a hexadecimal number: ").upper()
hexa = hexa[::-1]
hex_digits = "0123456789ABCDEF"
decimal = 0
for i in range(len(hexa)):
    value = hex_digits.index(hexa[i])
    decimal += value * (16 ** i)
print(decimal)
```

```python
# octal to binary
octal = input("Enter an octal number: ")
# Octal -> Decimal
octal_rev = octal[::-1]
decimal = 0
for i in range(len(octal_rev)):
    decimal += int(octal_rev[i]) * (8 ** i)
# Decimal -> Binary
binary = []
while decimal > 0:
    binary.append(str(decimal % 2))
    decimal = decimal // 2
binary = binary[::-1]
print("".join(binary))
```

```python
# hexadecimal to binary
hexa = input("Enter a hexadecimal number: ").upper()
hex_to_bin = {
    '0':'0000', '1':'0001',
    '2':'0010', '3':'0011',
    '4':'0100', '5':'0101',
    '6':'0110', '7':'0111',
    '8':'1000', '9':'1001',
    'A':'1010', 'B':'1011',
    'C':'1100', 'D':'1101',
    'E':'1110', 'F':'1111'
}
binary = ""
for digit in hexa:
    binary += hex_to_bin[digit]
print(binary.lstrip('0'))
```

```python
#binary to hex
binary = input("Enter binary number: ")

# pad to multiple of 4
while len(binary) % 4 != 0:
    binary = "0" + binary

hex_digits = {
    "0000":"0", "0001":"1", "0010":"2", "0011":"3",
    "0100":"4", "0101":"5", "0110":"6", "0111":"7",
    "1000":"8", "1001":"9", "1010":"A", "1011":"B",
    "1100":"C", "1101":"D", "1110":"E", "1111":"F"
}

hexa = ""

for i in range(0, len(binary), 4):
    chunk = binary[i:i+4]
    hexa += hex_digits[chunk]

print(hexa)
```

```python
def time(n):
    year = n // 365
    remaining_days = n % 365
    months = remaining_days // 30
    days = remaining_days % 30

    print(year, months, days)

n = int(input())
time(n)
```

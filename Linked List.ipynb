```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class SLL:
    def __init__(self):
        self.head = None

    # Insert at end
    def insert_end(self, data):
        new = Node(data)

        if self.head is None:
            self.head = new
            return

        temp = self.head

        while temp.next is not None:
            temp = temp.next

        temp.next = new

    # Traversal
    def traversal(self):
        if self.head is None:
            print("Linked List is Empty")
            return

        temp = self.head

        while temp is not None:
            print(temp.data, end=" -> ")
            temp = temp.next

        print("None")

    # Delete at end
    def delete_end(self):

        # Empty list
        if self.head is None:
            print("Linked List is Empty")
            return

        # Only one node
        if self.head.next is None:
            self.head = None
            return

        # Reach second last node
        temp = self.head

        while temp.next.next is not None:
            temp = temp.next

        # Delete last node
        temp.next = None


# Driver Code
obj = SLL()

obj.insert_end(10)
obj.insert_end(20)
obj.insert_end(30)
obj.insert_end(40)

print("Before deletion:")
obj.traversal()

obj.delete_end()

print("After deletion:")
obj.traversal()
```

```python
# Node Class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


# Singly Linked List Class
class SLL:

    def __init__(self):
        self.head = None

    # Insert at Beginning

    def insert_beginning(self, data):

        # Step 1: Create a new node
        new = Node(data)

        # Step 2: Point new node to current head
        new.next = self.head

        # Step 3: Make new node the new head
        self.head = new

        print(data, "inserted at beginning.")

    # Insert at End
    def insert_end(self, data):

        # Step 1: Create a new node
        new = Node(data)

        # Case 1: Linked List is empty
        if self.head is None:
            self.head = new
            print(data, "inserted as the first node.")
            return

        # Step 2: Traverse to the last node
        temp = self.head

        while temp.next is not None:
            temp = temp.next

        # Step 3: Link last node to new node
        temp.next = new

        print(data, "inserted at end.")

    # -------------------------
    # Traversal
    # -------------------------
    def traversal(self):

        if self.head is None:
            print("Linked List is Empty")
            return

        temp = self.head

        print("Linked List:", end=" ")

        while temp is not None:
            print(temp.data, end=" -> ")
            temp = temp.next

        print("None")

# Driver Code
obj = SLL()

# Insert at Beginning
obj.insert_beginning(30)
obj.insert_beginning(20)
obj.insert_beginning(10)

print("\nAfter inserting at beginning:")
obj.traversal()

# Insert at End
obj.insert_end(40)
obj.insert_end(50)

print("\nAfter inserting at end:")
obj.traversal()
```

```python
#Sum of nodes in a linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Singly Linked List class
class SLL:
    def __init__(self):
        self.head = None

    # Insert at end
    def insert_end(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
        else:
            temp = self.head
            while temp.next is not None:
                temp = temp.next
            temp.next = new_node
    # Sum of all nodes
    def sum_of_nodes(self):
        if self.head is None:
            return 0

        total = 0
        temp = self.head

        while temp is not None:
            total += temp.data      # Add current node value
            temp = temp.next        # Move to next node

        return total
# Driver Code
obj = SLL()
obj.insert_end(10)
obj.insert_end(20)
obj.insert_end(30)
obj.insert_end(40)

print("Sum of nodes =", obj.sum_of_nodes())
```

```python
#count no.of nodes
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Singly Linked List class
class SLL:
    def __init__(self):
        self.head = None

    # Insert at end
    def insert_end(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
        else:
            temp = self.head
            while temp.next is not None:
                temp = temp.next
            temp.next = new_node

    # Count number of nodes
    def count_nodes(self):
        count = 0
        temp = self.head

        while temp is not None:
            count += 1          # Count this node
            temp = temp.next    # Move to next node

        return count
obj = SLL()

obj.insert_end(10)
obj.insert_end(20)
obj.insert_end(30)
obj.insert_end(40)
obj.insert_end(50)
print("Number of nodes =", obj.count_nodes())
```

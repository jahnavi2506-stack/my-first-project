```python
def length(s):
    words = s.split()
    return len(words[-1])
s = input()
print(length(s))
```

```python
#remove duplicates using set
def remove_duplicate(s):
    seen = set()
    result = []
    for ch in s:
        if ch not in seen:
            seen.add(ch)
            result.append(ch)
    return "".join(result) 
s = input()    
print(remove_duplicate(s))
```

```python
#remove duplicates
def remove_dup(s):
    s2 = " "
    for ch in s:
        if ch not in s2:
            s2 += ch
    return s2
s=input()
print(remove_dup(s))
```

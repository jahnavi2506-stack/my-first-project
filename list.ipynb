```python
def duplicate(nums):
    return len(nums) != len(set(nums))
nums = list(map(int,input().split()))
print(duplicate(nums))
```

```python
nums = [1, 1, 2, 2, 3, 4, 4, 5]

unique = []

for num in nums:
    if num not in unique:
        unique.append(num)

print(unique)
```

```python
#print even number in desc and odd number in asc from selected number
n = int(input())

for i in range(n, -1, -1):
    if i % 2 == 0:
        print(i, end=" ")
print()

for i in range(1, n + 1):
    if i % 2 != 0:
        print(i, end=" ")
        
```

```python
arr = list(map(int, input().split()))
k = int(input())
sum = sum(arr[:k])
max_sum = sum
for i in range(k, len(arr)):
    sum = sum -arr[i - k] + arr[i]
    max_sum = max(max_sum, sum)
print(max_sum)    
#algo in vs code    
```

```python
distance = int(input()) 
bookingHour = int(input())
fare = 100.0
if distance > 10:
    fare += (distance - 10) * 10
if bookingHour >= 22 or bookingHour <= 6:
    fare += 50
print(f"{fare:.2f}")
```

```python
#printing the greater prime number 
n = int(input())
for num in range(n, 1, -1):
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num)
        break
else:
    print(-1)
```

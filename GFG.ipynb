```python
d = dict() 
for x in enumerate(range(2)): 
	d[x[0]] = x[1] 
	d[x[1]+7] = x[0] 
print(d)
```

```python
d = {} 

def addTodict(country): 
	if country in d: 
		d[country] += 1
	else: 
		d[country] = 1

addTodict('China') 
addTodict('Japan') 
addTodict('china') 
print(d)
```

```python
#reverse of array using two-pointers
arr = list(map(int, input().split()))
left = 0
right = len(arr) - 1
while left < right:
    arr[left], arr[right] = arr[right], arr[left]
    left += 1
    right -= 1
print(arr)
```

```python
#reverse of a string palindrome
s = input()
left = 0
right = len(s) - 1
while left < right:
    if s[left] != s[right]:
        print("Not Palindrome")
        break
    left += 1
    right -= 1
else:
    print("Palindrome")
```
